### Rolling mean
The above shows the daily variation, which is useful  to understand. However, to show long-term trends, i'll use a rolling mean:

except from the first day (which has no previous day), i'll take the **mean** of the value of the day + value of the previous day 

In python:
values['rolling_mean'] = values['daily_values'].rolling(2).mean()
- rolling window parameter, rolling() if 2, then mean between current and previous. if 3, current and previous 3, etc.
- some common values include seven (one week), 30 (one month), 50 or 100 days, or 365 days (one year)

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
```

```python
euro_to_dollar['rolling_mean']
```

### Using a for loop .zip() method to create a few different rolling means

```python
plt.figure(figsize=(13, 8))

for i, rolling_mean in zip([1, 2, 3, 4], [7, 30, 100, 365]):
    
    plt.subplot(2, 2, i)
    plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('($) Rolling Window:' + str(rolling_mean))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(13, 8))

for i, rolling_mean in zip([1, 2, 3, 4], [7, 30, 100, 365]):
    
    plt.subplot(2, 2, i)
    plt.plot(euro_to_gbp['Time'], euro_to_gbp['UK_pound_sterling'].rolling(rolling_mean).mean())
    plt.title('(Â£) Rolling Window:' + str(rolling_mean))
plt.tight_layout()
plt.show()
```
